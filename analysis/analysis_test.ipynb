```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm
```

```python
path = "/scratch/00993/tg802402/NNrealgeographyonline001_3epoch/run"
#path = "/scratch/00993/tg802402/NNrealgeographyonline001/run"
path0 = "/scratch/00993/tg802402/CTRLrealgeographyonline001/run"
# first sim-day:
files = xr.open_mfdataset(path+"/*.cam.h1.2004-01-01-*.nc")
files0 = xr.open_mfdataset(path0+"/*.cam.h1.2004-01-01-*.nc")
day1_T_zmean= files.TBP.mean(dim=("lon","time")).values
day1_T0_zmean= files0.TBP.mean(dim=("lon","time")).values
```

```python
plt.figure(figsize=(15,10))
plt.subplot(311)
plt.imshow(day1_T0_zmean)
plt.colorbar()
plt.title ('CTRL')
plt.subplot(312)
plt.imshow(day1_T_zmean)
plt.colorbar()
plt.title ('TEST')
plt.subplot(313)
anomaly = day1_T_zmean-day1_T0_zmean
plt.imshow(anomaly,norm=TwoSlopeNorm(0), cmap=plt.cm.seismic)
plt.colorbar()
plt.title ('TEST-CTRL')

```
